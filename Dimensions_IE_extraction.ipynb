### Patent

```python
# *EP.csv로 끝나는 파일은 2010년 이후의 데이터를 포함하고 있음
# *EP.csv로 끝나는 모든 파일을 pat_data_df_all 이라는 이름의 데이터프레임으로 합치기
pat_data_df_all = pd.DataFrame()
for i in range(2010, 2020):
    pat_data_df = pd.read_csv(dir + "pat_data_df_" + str(i) + "_EP.csv", encoding = "ISO-8859-1")
    pat_data_df_all = pd.concat([pat_data_df_all, pat_data_df], axis = 0)
```

### Publication

```python
# publication 폴더 안에 있는 pub_data_df 이름으로 시작하는 모든 파일 pub_data_df_all이라는 데이터프레임으로 합치기
pub_affiliations_df_all = pd.DataFrame()
for i in range(2000, 2020):
    pub_affiliations_df = pd.read_csv(dir + "publication/pub_affiliations_df_" + str(i) + "_EP_1.csv", encoding = "ISO-8859-1")
    pub_affiliations_df_all = pd.concat([pub_affiliations_df_all, pub_affiliations_df], axis = 0)    
```

```python
pub_affiliations_df_all.pub_id.nunique()
```

```python
pub_affiliations_df_IE = pub_affiliations_df_all[pub_affiliations_df_all.country_code == 'IE'].drop_duplicates()
pub_affiliations_df_IE.pub_id.nunique()
```

```python
pub_affiliations_df_IE.head()
```

```python
pub_affiliations_df_IE.to_csv(dir + "IE/pub_affiliations_df_IE.csv", index = False)
```

```python
# publication 폴더 안에 있는 pub_data_df 이름으로 시작하는 모든 파일 pub_data_df_all이라는 데이터프레임으로 합치기
pub_data_df_all = pd.DataFrame()
for i in range(2000, 2020):
    pub_data_df = pd.read_csv(dir + "publication/pub_data_df_" + str(i) + "_EP_1.csv", encoding = "ISO-8859-1")
    pub_data_df_all = pd.concat([pub_data_df_all, pub_data_df], axis = 0)    
```

```python
pub_data_df_all.head()
```

```python
# pub_data_df_all 데이터프레임과 pub_affiliations_df_all 데이터프레임을 pub_data_df_all 데이터프레임을 기준으로 병합
pub_data_df_IE = pd.merge(pub_data_df_all, pub_affiliations_df_IE[['pub_id','country','country_code']], how = 'inner', on = 'pub_id')
pub_data_df_IE
```

```python
pub_data_df_IE.to_csv(dir + "IE/pub_data_df_IE.csv", index = False)
```
